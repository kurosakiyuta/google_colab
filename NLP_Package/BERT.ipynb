```
#text = "私はサッカー選手です。そして、テニスもします。加えて、バトミントンが好きです。"
def mecab_tokenizer(text):
    result = mecab.parse(text)
    tokenized_text = [r.replace('。', '[SEP]') for r in result.replace('\n', '').split(' ') if r != ''  ]
    tokenized_text.insert(0, '[CLS]')
    tokenized_text_limitation = tokenized_text[:511]
    if tokenized_text_limitation[-1] != '[SEP]':
        tokenized_text_limitation.append('[SEP]')
    return tokenized_text_limitation

```

```
#device = xm.xla_device()
device
def compute_vector(text, model = bert_model, bert_tokenizer = bert_tokenizer, mecab_tokenizer= mecab_tokenizer):
    use_model = model
    tokens = mecab_tokenizer(text)
    #print(tokens)
    token_ids = bert_tokenizer.convert_tokens_to_ids(tokens)
    tokens_tensor = torch.tensor(token_ids).unsqueeze(0)
   # print(tokens_tensor.size())
    use_model.eval()
    tokens_tensor = tokens_tensor.to(device)
    model.to(device)
    with torch.no_grad():
        outputs = use_model(tokens_tensor)
        last_hidden_states = outputs[0]
#    pooling_layer = -2
    embedding = last_hidden_states#all_encoder_layers[0][pooling_layer].numpy()
    # embedding = all_encoder_layers[0].numpy()
    # return np.mean(embedding, axis=0)
    return embedding
# embedding is text embedding
```

```
import numpy as np

def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```

def suisen(df, emb):
    sim = []
    for i in range(len(df)):
        sim.append([df.iloc[i,0], cos_sim(list(df.iloc[i, 1:]), emb)])
    sim_sort = sorted(sim, key=lambda x: x[1], reverse=True)
    return sim_sort
```
